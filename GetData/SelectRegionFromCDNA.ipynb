```python
from Bio import SeqIO
import glob

all = glob.glob("mart_export*.txt")

with open("Sec1.fasta","w") as f1,open("Sec2.fasta","w") as f2,open("Sec3.fasta","w") as f3,open("Sec4.fasta","w") as f4,open("Sec5.fasta","w") as f5,open("Sec6.fasta","w") as f6,open("Sec7.fasta","w") as f7,open("Sec8.fasta","w") as f8:
    for file_ in all:
        for seq_record in SeqIO.parse(file_, "fasta"):      
            i = 1
            while (i < 9):
                if i==1:
                    f8.write(">" + str(seq_record.id) + "\n")
                    f8.write(str(seq_record.seq[-50:] + "\n"))                 
                if i==2:
                    f7.write(">" + str(seq_record.id) + "\n")
                    f7.write(str(seq_record.seq[-50*i:-50*(i-1)] + "\n"))
                if i==3:
                    f6.write(">" + str(seq_record.id) + "\n")
                    f6.write(str(seq_record.seq[-50*i:-50*(i-1)] + "\n"))
                if i==4:
                    f5.write(">" + str(seq_record.id) + "\n")
                    f5.write(str(seq_record.seq[-50*i:-50*(i-1)] + "\n"))
                if i==5:
                    f4.write(">" + str(seq_record.id) + "\n")
                    f4.write(str(seq_record.seq[-50*i:-50*(i-1)] + "\n"))
                if i==6:
                    f3.write(">" + str(seq_record.id) + "\n")
                    f3.write(str(seq_record.seq[-50*i:-50*(i-1)] + "\n"))
                if i==7:
                    f2.write(">" + str(seq_record.id) + "\n")
                    f2.write(str(seq_record.seq[-50*i:-50*(i-1)] + "\n"))
                if i==8:
                    f1.write(">" + str(seq_record.id) + "\n")
                    f1.write(str(seq_record.seq[-50*i:-50*(i-1)] + "\n"))

                i += 1
              
```
